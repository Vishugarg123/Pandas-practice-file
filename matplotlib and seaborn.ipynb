```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4,5,6])
```

```python
b=np.array([23,45,6,45,78,90])
```

```python
plt.title('First graph')
plt.xlabel(a)
plt.ylabel(b)
plt.plot(a,b)
plt.show()
```

```python
g=np.array([23,34,45,56,67,78,89,90,10])
h=np.array([23,34,43,54,65,76,87,98,90])
plt.title("Graph")
plt.xlabel(g)
plt.ylabel(h)
plt.plot(g,h,"--y")
plt.show()
```

```python
plt.title("bar graph")
plt.xlabel(g)
plt.ylabel(h)
plt.bar(g,h)
plt.show()
```

```python
plt.title("bar graph")
plt.xlabel(g)
plt.ylabel(h)
plt.bar(g,h,color='r',align="center")
plt.show()
```
